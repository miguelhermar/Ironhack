```python
import warnings
warnings.simplefilter('ignore') 
```

```python
import requests
import pandas as pd
import time
from pandas.io.json import json_normalize
```

```python
import re
```

```python
response = requests.get('https://api.fbi.gov/artcrimes?page=1').json()
```

```python
response
```

```python
response.keys()
```

```python
flattened_data = json_normalize(response['items'])

flattened_data.drop(['uid', 'url', 'idInAgency', 'images', 'path'], axis=1, inplace=True)
```

```python
flattened_data
```

```python
total_pages = response['total']
total_pages
```

```python
df=pd.DataFrame()
k = 1
for i in range(total_pages):
    url= f'https://api.fbi.gov/artcrimes?page={k}'
    response = requests.get(url)
    results = response.json() 
    
    flattened_data = json_normalize(results['items'])
    
    df = df.append(flattened_data)
    k += 1
    if i == 100:
        break
```

```python
df
```

```python
df.reset_index(inplace=True)
```

```python
df.drop('index', axis=1, inplace=True)
```

```python
df.drop(['path', 'uid', 'url', 'idInAgency', 'images'], axis=1, inplace=True)
```

```python
df.to_csv('Art Theft API', encoding='utf-8')
```

### Web Scraping

```python
import requests
from bs4 import BeautifulSoup
```

```python
def init_bs(url):
    webpage = requests.get(url).content
    return BeautifulSoup(webpage, "html")
```

```python
tyree = init_bs('https://www.tyreeguyton.com/about')
```

```python
tyree
```

```python
exhib = tyree.find_all('ul')[4]
exhib = [i.text for i in exhib.find_all('p')]
exhib
```

```python
place = [re.findall(r'[A-z]+?,? ?[A-z]+[A-z]+$', i) for i in exhib]
```

```python
place
```

```python
year = [elem[:4] for elem in exhib]
```

```python
text = [re.findall(r'[A-z]+', i) for i in exhib]
del text[0][0]
```

```python
text = [" ".join(i) for i in text]
```

```python
tyree_df = pd.DataFrame([year, text, place]).transpose()
tyree_df.rename(columns={0: 'Year', 1: 'Piece Name', 2: 'Location'}, inplace= True)
```

```python
tyree_df
```

```python
tyree_df = tyree_df[tyree_df['Piece Name'].str.contains('Faces in the Hood')].reset_index()
```

```python
tyree_df.drop('index', axis=1, inplace=True)
tyree_df.drop('Piece Name', axis=1, inplace=True)
```

```python
tyree_df
```

```python
titles = [['Faces in the Hood Series, Faces of God Series'], 'Lincoln in Dalivision',  
          'Cosmographia', ['Human Face Pendant', 'Lizard Pendant'], 'Saint Anthony Miracles', 'Untitled Abstract Painting', 
          'Prajnaparamita Sutra in 600 Chapters',] 
```

```python
dali = init_bs('https://en.wikipedia.org/wiki/Lincoln_in_Dalivision')
```

```python
dali
```

```python
infobox = dali.find(class_='infobox vevent')
rows = [row.text.replace('\n', ' ').replace('Artist', '').replace('Location', '') for row in infobox.find_all('tr')]
rows = rows[2:]
```

```python
rows
```

```python
dali_df = pd.DataFrame([rows])
```

```python
dali_df.rename(columns={0: 'Year', 1: 'Location'}, inplace= True)
```

```python
dali_df
```

```python
ptolom = init_bs('https://www.themorgan.org/exhibitions/images/ulm-ptolemy')
```

```python
ptolom
```

```python
cosmos = ptolom.select('div.field-item.even')[1]
cosmos = [i.text for i in cosmos.find_all('p')]
cosmos = [j for i in cosmos for j in i][-36:]
```

```python
cosmos_place = "".join(cosmos)
cosmos_place
```

```python
cosmos = ptolom.select('div.field-item.even')[1]
cosmos = [i.text for i in cosmos.find_all('p')]
cosmos = [j for i in cosmos for j in i][61:65]
cosmos_date = "".join(cosmos)
cosmos_date
```

```python
ptolom_df = pd.DataFrame([cosmos_date, cosmos_place]).transpose()
ptolom_df
```

```python
ptolom_df.rename(columns={0: 'Year', 1: 'Location'}, inplace= True)
```

```python
ptolom_df
```

```python
df_o = pd.concat([tyree_df,dali_df, ptolom_df], ignore_index = True, axis = 0)

df_o
```
